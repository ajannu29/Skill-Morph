```python
n = int(input("Enter a number: "))

print(f"Even numbers between 1 and {n} are:")

for i in range(1, n + 1):
    if i % 2 == 0:
        print(i, end=" ")
```

```python
n = int(input("Enter a number: "))

total = 0

for i in range(1, n + 1):
    if i % 3 == 0 or i % 5 == 0:
        total += i

print("sum of numbers divisible by 3 or 5:", total)
```

```python
n = int(input("Enter a number: "))

print(f"Multiplication table of {n}:")

for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")
```
